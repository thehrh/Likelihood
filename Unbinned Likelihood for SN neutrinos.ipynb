### Double integration test

```python
from scipy.integrate import dblquad, quad
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(y,x):
    return norm.pdf(x)*norm.pdf(y)

myf = np.vectorize(f)
```

```python
xs,ys = np.meshgrid(np.linspace(-3,3,100),np.linspace(-3,3,100))
```

```python
res = myf(ys,xs)
```

```python
plt.imshow(res,extent=[-3,3,-3,3],origin='lower')
```

### Integration testen

```python
dblquad(myf,-3,3,lambda x: -3, lambda x: 3)
```

```python
intgral = quad(norm.pdf,-3,3)[0]
```

### Funktion

Energiefaltung: 
$$ P(E_{\rm reco}) = \int R(E_{\rm reco} - E_{\rm true}) P(E_{\rm true}) \, dE_{\rm true} $$

Zeitfaltung:
$$ P(t_{\rm reco}) = \int F(N, t_{\rm true} + t_{true})P(t_{\rm true}) \,dt_{\rm true} $$

Gesamtintegral:
$$ P(E_{\rm reco},t_{\rm reco})= \iint R(E_\rm{reco} - E_{\rm true}) F(N,t_{\rm true} + t_{\rm true}) P(m_\nu,E_{\rm true},t_{\rm true}) \, dE_{\rm true} dt_{\rm true}$$
